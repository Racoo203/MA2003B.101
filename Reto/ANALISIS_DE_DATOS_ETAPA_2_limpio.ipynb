## Librerias y datos limpios

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datos = pd.read_csv('dataset.csv')
datos.head()
```

```python
print("Datos totales")
print(datos.shape)
```

## Tipo de variables

```python
print("tipo de variables: ")
print(datos.dtypes)
print("descripción estadistica de las variables: ")
print(datos.describe(include='all'))
```

## Análisis estadístico de variables numericas

```python
COLUMNAS = datos.select_dtypes(include=['float64', 'int64']).columns

for col in COLUMNAS:
    print(f"{col.upper()}: ")
    print("MAX: ", datos[col].max())
    print("MIN: ", datos[col].min())
    print("Rango: ", datos[col].max() - datos[col].min())
    print("Varianza: ", datos[col].var())
    print("Desviación estándar: ", datos[col].std())
    print("Cuartiles: ")
    print(datos[col].quantile([0.25, 0.5, 0.75]))
    print("\n")
```

## Gráficas de variables numericas

```python
for columna in COLUMNAS:
    plt.figure(figsize=(8, 6))
    plt.hist(datos[columna])
    plt.title(f'{columna}')
    plt.xlabel(columna)
    plt.ylabel('Frecuencia')
    plt.grid(True)
    plt.show()
```

```python
for columna in COLUMNAS:
    plt.figure(figsize=(8, 6))
    plt.boxplot(datos[columna])
    plt.title(f'{columna}')
    plt.xlabel(columna)
    plt.ylabel('Frecuencia')
    plt.grid(True)
    plt.show()
```

## Aálisis de datos "Date" y "Clave_Estacion"

```python
print(datos['Date'].min())
print(datos['Date'].max())
```

```python
datos['Date'] = pd.to_datetime(datos['Date'])
DIAS = datos['Date'].dt.date.value_counts()
DIAS.plot(kind='bar')
DIAS
```

```python
datos['Year'] = datos['Date'].dt.year
data_by_year = datos['Year'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(data_by_year, labels=data_by_year.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.show()
```

```python
ESTACION = datos['Clave_Estacion'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(ESTACION, labels=ESTACION.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.show()
```

## Correlación entre variables

```python
datos2 = datos.select_dtypes(include='number')
plt.figure(figsize=(8, 7))
sns.heatmap(datos2.corr(), annot=True, cmap='coolwarm', fmt='.1f', cbar=True, vmin=-1, vmax=1)
plt.show()
```

## Datos nulos, atipicos y repetidos

```python
datos.isnull().sum()
```

```python
outliers_counts = {}
for column in datos.columns:
    if datos[column].dtype != 'object':
        mean = datos[column].mean()
        std = datos[column].std()
        lower_bound = mean - 5 * std
        upper_bound = mean + 5 * std
        outliers = datos[(datos[column] < lower_bound) | (datos[column] > upper_bound)]
        outliers_count = len(outliers)
        outliers_counts[column] = outliers_count
        print(f'{column}: {outliers_count}')
```
